```python
# 匯入套件
from bs4 import BeautifulSoup as bs
import requests as req
from pprint import pprint

# 操作 browser 的 API
from selenium.webdriver.chrome.service import Service
from selenium import webdriver

# 處理逾時例外的工具
from selenium.common.exceptions import TimeoutException

# 面對動態網頁，等待某個元素出現的工具，通常與 exptected_conditions 搭配
from selenium.webdriver.support.ui import WebDriverWait

# 搭配 WebDriverWait 使用，對元素狀態的一種期待條件，若條件發生，則等待結束，往下一行執行
from selenium.webdriver.support import expected_conditions as EC

# 期待元素出現要透過什麼方式指定，通常與 EC、WebDriverWait 一起使用
from selenium.webdriver.common.by import By

from selenium.webdriver.common.keys import Keys

# 強制等待 (執行期間休息一下)
from time import sleep

import os, json,re


# 啟動瀏覽器工具的選項
my_options = webdriver.ChromeOptions()
# my_options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
my_options.add_argument("--start-maximized")         #最大化視窗
my_options.add_argument("--incognito")               #開啟無痕模式
my_options.add_argument("--disable-popup-blocking") #禁用彈出攔截
my_options.add_argument("--disable-notifications")  #取消 chrome 推播通知
my_options.add_argument("--lang=zh-TW")  #設定為正體中文


# 使用 Chrome 的 WebDriver
my_service = Service(executable_path="./chromedriver.exe")
driver = webdriver.Chrome(
    options = my_options,
    service = my_service
)
```

```python
# 取得新聞列表
url = "https://www.gutenberg.org/browse/languages/zh" 

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url) 

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml") 
```

```python
# 建立 list 來放置列表資訊
list_books = []

# 取得 列表 的文字與超連結，要有href a屬性的我才要div.title也可以省略，直接找最裡面的連結
#for a in soup.select('li.pgdbetext a[href]'):

#取得li class的pgdbetext
for style in soup.select('li.pgdbetext'):
    
    #取得a的連結格式
    a=style.a
    
    #判斷內文裡面不要有English 的字 且 re套件裡面找到標題要有中文
    if(style.text.find("English")!=-1 or bool(re.findall(r"[\u4e00-\u9fff]+",a.text))==False):
        #print(a.text+" pass")
        continue
        
       
                    
    #https://www.gutenberg.org/files/25328/25328-0.txt   
    #link = a['href']
    #print(link)
    #把目錄的id切分出來， print(id[2]) 指顯示目錄值
    #id = link.split("/") # 以/切link 第一個 /後是0 第二個是1 ...
    
        
    
    id=a['href'].split("/")[2]
    print(a.text)
        #if find_elements_by_class.pgdbetext !=  (English) (as Author):

    try:
           # 加入列表資訊
            list_books.append({
                   'title': a.get_text(),
                   'link': 'https://www.gutenberg.org' + a['href'],
                    'downloadLink': 'https://www.gutenberg.org/files/'+id+'/'+id+'-0.txt' 
                })
    except Exception as e: 
            print(e)


    #print(a['href']) 
    #print(list_books[0]['title'])
    #print(list_books[0]['link'])
    #print(list_books[0]['downloadLink'])
```

```python
folderPath = 'ebook'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)
```

```python
#總書目計算
list_books_index=1
count = 1 
while list_books_index<=5: 
    
    driver.get(list_books[list_books_index]['downloadLink'])
   
   
  
    #找尋網頁tag底下的內文用content作為變數
    content=driver.find_element(By.TAG_NAME, "pre")
    
    regex=r"[a-zA-Z\s\d*:,.\-#\[\]]"
    content_Chinese= re.sub(regex,'',content.text)
    #print(content_Chinese)
    
    
    
    #判斷書名title在內文裡
    if ((list_books[list_books_index]['title'] in content_Chinese) == True):
        print(list_books[list_books_index]['title'])  
            
        #print(content.text.split(list_books[list_books_index]['title'])[1])
        print(os.path.abspath(folderPath))
        
        
        try:
            x = os.path.join(os.path.abspath(folderPath), f"第{count}本_{list_books[list_books_index]['title']}.txt")
            with open(x,'w',encoding='UTF-8') as f:
                f.write(content_Chinese)
                f.close()
                
        except Exception as e: 
            print(e)

        print('第'+str(count)+'本_'+list_books[list_books_index]['title'])
        list_books_index+=1 
        count+=1
    else: 
        print('不符合'+ str(list_books_index))
        list_books_index+=1
```
